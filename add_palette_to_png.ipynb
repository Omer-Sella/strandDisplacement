# Add PLTE to PNG

Open a PNG, convert it to indexed palette mode, and save a new PNG containing a `PLTE` chunk.

```python
from pathlib import Path
from PIL import Image


def add_plte_to_png(input_path: str, output_path: str, colors: int = 256):
    """Convert PNG to indexed palette mode, which writes a PLTE chunk."""
    if colors < 1 or colors > 256:
        raise ValueError('colors must be in the range [1, 256].')

    in_path = Path(input_path)
    out_path = Path(output_path)

    if not in_path.exists():
        raise FileNotFoundError(f'Input PNG not found: {in_path}')

    img = Image.open(in_path).convert('RGBA')

    # Quantize to palette mode; saving this image writes PLTE.
    pal_img = img.convert('P', palette=Image.ADAPTIVE, colors=colors)
    pal_img.save(out_path, format='PNG')

    return out_path
```

```python
# Example usage
input_png = 'imperialBlue.png'
output_png = 'imperialBlue_with_plte.png'

saved_path = add_plte_to_png(input_png, output_png, colors=256)
print(f'Saved PNG with PLTE at: {saved_path.resolve()}')
```
