```python

from nupack import pfunc
from nupack import mfe, Complex, Strand, Model

from nupack import * 
dnaLibrary = ['AAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAA', 'GGGGTTTTTTTTTTTTTTTTTTTTTTTTTTTTT']
nextCandidateDNA = 'CGCGCGCGCGCG'
# Define a NUPACK tube that contains all the strands in the dna library at 1 µM, AND the nextCandidateDNA
dnaLibraryAsTubeDictionary = {Strand(string=s, name=f'strand_{i}'): 1e-6 for i, s in enumerate(dnaLibrary)}
dnaLibraryAsTubeDictionary[Strand(string=nextCandidateDNA, name='candidate')] = 1e-6
model = Model(material='dna', celsius=37, sodium=0.05)
tube = Tube(
    dnaLibraryAsTubeDictionary,
    complexes=SetSpec(max_size=2),   # consider monomers + all 2-strand complexes
    name='test_tube'
)

result = tube_analysis(
    tubes=[tube],
    model=model,
    compute=['pfunc', 'mfe', 'pairs']
)
# Print all complexes sorted by concentration
concentrations = result[tube].complex_concentrations
sorted_complexes = sorted(concentrations.items(), key=lambda x: float(x[1]), reverse=True)

print(f"{'Complex':<25} {'ΔG (kcal/mol)':>15} {'Conc (M)':>15}")
print("-" * 55)
for complex_, conc in sorted_complexes:
    strand_names = '+'.join(s.name for s in complex_.strands)
    dG = result[complex_].free_energy
    print(f"{strand_names:<25} {dG:>15.2f} {float(conc):>15.2e}")
```
